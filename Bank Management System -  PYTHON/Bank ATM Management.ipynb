```python
class Bank:
    def __init__(self,AccountList):
        self.AccountList = AccountList
        
    def display(self):
        for i in self.AccountList:
            print(i)
            
    def validateAccNo(self,flag,userAcc):
        self.flag = flag
        self.userAcc = userAcc
        if(self.userAcc in self.AccountList):
            print("User Exits \n ")
            self.flag =1
            return self.flag,self.userAcc
        else:
            print("User doesn't exist ")
            return self.flag ,self.userAcc
        
        
class CutomerDetails:
    pass
    def __init__(self,CustomerDetails):
        self.CustomerDetails = CustomerDetails
        #for i in self.CustomerDetails:
        #    print(i)
        
    def atmServices(self,flag,userAcc):
        self.flag    = flag
        self.userAcc = userAcc
        self.log = 0
        if(self.flag == 1 ):
            self.pin = int(input("Please enter Your ATM PIN : "))
            for i in range(len(self.CustomerDetails)):                   
                        if(self.pin == self.CustomerDetails[i][3]):
                            print("Logged in Successfully ")
                            self.log = 1
                            break
            if(self.log==1):
                ch = 0
                while(ch!=4):
                    print("\n")
                    print("""          **** WELCOME TO ATM SERVICES ****
                    PRESS 1 : BALANCE ENQUIRY
                    PRESS 2 : CASH DEPOSIT
                    PRESS 3 : CASH WITHDRAW
                    PRESS 4 : EXIT 
                    """)
                    ch = int(input("Pleas Enter Your Choice : "))
                
                    if(ch==1):
                        for i in range(len(self.CustomerDetails)):
                            if(self.userAcc == self.CustomerDetails[i][0]):
                                    print("Your Balance is  : {}".format(self.CustomerDetails[i][2]))
                                
                    elif(ch==2):
                        self.depo = int(input("Please Enter Deposit Amount "))
                        for i in range(len(self.CustomerDetails)):
                            if(self.userAcc == self.CustomerDetails[i][0]):
                                self.CustomerDetails[i][2]+=  self.depo
                                print("Case Deposited Successfully")
                                
                    elif(ch==3):
                        self.a = int(input("Please Enter withdrawal Amount "))  
                        for i in range(len(self.CustomerDetails)):
                            self.rem  = self.CustomerDetails[i][2] - self.a
                            
                            if(self.userAcc == self.CustomerDetails[i][0]):
                                if(self.a > self.CustomerDetails[i][2] ):
                                    print("Please Enter a valid Amount ")
                                    break
                                elif(self.rem <500):
                                    print("Transaction can't Be Processed ")
                                    break
                                else: 
                                    self.CustomerDetails[i][2]-=  self.a
                                    break                                     
                     
                              
                    elif(ch==4):
                        print("Thank You Visit Again ! ")
                        break
                                    
            else:
                print("You Have Entered the wrong  ATM PIN ")
                                
        else:
            print("You Have Entered the wrong Account Number ")
                    
                    
                
                
def main():   
    B = Bank([123,456,789])
    Cust = CutomerDetails([[123,"Vinod",1000,111],[456,"Akash",3000,222],[789,"Sunny",50000,333]])
    flag,userAcc = B.validateAccNo(0,int(input("Please Enter Your Account Number : ")))
    Cust.atmServices(flag,userAcc)
    #B.validateAccNo(0,int(input("Please Enter Your Account Number : ")))
main()
```
