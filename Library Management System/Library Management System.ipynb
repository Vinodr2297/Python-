```python
class Library:
    BookNames = []
    def __init__(self,BooksNames):
        Library.BookNames = BooksNames
    
    def displayBookNames(self):
        print("\n The Books we have in our library is : \n")
        if(len(Library.BookNames)>0):
            for i in Library.BookNames:
                print(i)
            print("\n ")
        else:
            print("Currently No Books are Avialable in Library : ")
            
    def checkAvailablity(self,reqBook):
        self.reqBook = reqBook
        if(self.reqBook in Library.BookNames):
            Library.BookNames.remove(self.reqBook)
            print("You have now Borrowed a book called {} \n ".format(self.reqBook))
        else:
            print("Sorry the Requested Book is not available \n  ")
    
    def addBook(self,addBook):
        self.addBook = addBook
        Library.BookNames.append(self.addBook)

class Student:
    def requestBook(self):
        self.reqBook = input("Enter the name of the book you want : ")
        return self.reqBook
    
    def addBook(self):
        self.addBook  = input("Enter the Name of the returning Book")
        return self.addBook

def main():
    Lib  = Library(["Lost of Us","Dark Knight rises","The King and Queen"])
    stud = Student() 
    choice  = 0
    while(choice!=4):
        print("""\n                       *_*_*_*_ LIBRARY MANAGEMENT _*_*_*_*

                    press 1 . Display all available Books 
                    press 2 . Request a Book
                    press 3 . Return  a Book
                    press 4 . Exit
             """ )
        choice = int(input(("Enter Your Choice :")))
        if(choice == 1):
            Lib.displayBookNames()
            
        elif(choice == 2):
            Lib.checkAvailablity(stud.requestBook())
        
        elif(choice==3):
            Lib.addBook(stud.addBook())
        elif(choice == 4):
            print("Thank You Happy reading ")
            break 
            
main()      
```
